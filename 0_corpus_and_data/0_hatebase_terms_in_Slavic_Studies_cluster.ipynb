```python
import pandas as pd
from pathlib import Path
main_dict = {}
```

```python
from tqdm import tqdm 

ngram1 = Path('slavic_review_data/SlavicStudiesCluster1991to2020/ngram1')
for file in tqdm(ngram1.iterdir(), total=len(list(ngram1.iterdir()))):
    text = file.read_text()
    for line in text.split('\n'):
        word=line.split('\t')[0]
        freq=line.split('\t')[-1]
        if freq == '':
            pass
        else:
            freq = int(freq)
            assert isinstance(freq, int)

        if word in main_dict.keys():
            main_dict[word] = main_dict[word] + freq
        else:
            main_dict[word] = freq
            
        
    
```
